```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Loading Data 
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

# Pie Chart

```python
df.head(4)
```

```python
df['Remote'] = df['job_work_from_home'].apply(lambda x: 'Yes' if x == True else 'No')
```

```python
df.head(4)
```

```python
df['Remote'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')  # Startangle determines what angle the pie slice begins. autopct is for formatting the percentage. f specifies it's going to be a float. The two %% after specifies we need the % symbol

plt.title('Work From Home Status')
plt.ylabel('')
plt.show()
```

```python
# Let's do this for other columns with boolean values

df['Degree_mentioned'] = df['job_no_degree_mention'].apply(lambda x: 'Yes' if x == True else 'No')
df['Health_Insurance_Available'] = df['job_health_insurance'].apply(lambda x: 'Yes' if x == True else 'No')
```

Using matplotlib methods to plot this time
We can plot them individually but it is cooler and less cumbersome to code it using a for loop. For this we would assign a dict to a variable. The keys would be the columns and the items would be the title.

```python
fig, ax = plt.subplots(1, 3) # These specify the row and the column

pie_dict = {
    'Remote': 'Work From Home',
    'Degree_mentioned': 'Job Degree Req.',
    'Health_Insurance_Available': 'Health Ins. Offered'
}

for i, (column, title) in enumerate(pie_dict.items()):  # i specifies the index and since we are working with a dict, we have to put our keys and items in a tuple otherwise we would get an error
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['No', 'Yes'])
    ax[i].set_title(title)

plt.show()
```

## Scatter Plot

```python
ds_df = df[df['job_title_short'] == 'Data Scientist'].copy()
```

```python
ds_exploded = ds_df.explode('job_skills')

# With the aggregate function we created a new column and with an embedded tuple specified the column to aggregate and the operation to do on them.
ds_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)
```

```python
# Assigning to a variable then sort by descending order.
skill_stat = ds_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)
```

```python
skill_stat = skill_stat.sort_values(by='skill_count', ascending=False).head(10)
```

```python
skill_stat.plot(kind= 'scatter', x='skill_count', y='median_salary')
plt.xlabel('Count of Job Posting')
plt.ylabel('Median Yearly Salary (USD)')
plt.title('Salary vs. Count of Job Postings for Top 10 Skills')
plt.tight_layout()
plt.show()
```
