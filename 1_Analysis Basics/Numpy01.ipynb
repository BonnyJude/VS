```python
# Install Datasets Library
!pip install datasets

# Import Libraries
from datasets import load_dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
# Converting datasets into a Panda dataframe from a train
df = dataset['train'].to_pandas()
```

```python
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Exercise!!
# Visualising the median salary for each job
```

```python
# Grouping by salary and sorting
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
# Plotting with Pandas and matplotlib

job_salary.plot(kind='barh')
plt.xlabel('Average Salary in USD')
plt.ylabel('')
plt.title('Average Salary for Each Job')
plt.show()
```
