```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# Loading Data 
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
```

```python
df_usa
```

```python

df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')

df_usa
```

```python
df_usa.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')
```

```python
df_usa_pivot = df_usa.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')

df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace = True)
```

```python
df_usa_pivot
```

```python
df_usa_pivot['Month Nos'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format = '%B').dt.month

df_usa_pivot
```

```python
df_usa_pivot.sort_values('Month Nos', inplace = True)

df_usa_pivot
```

```python
df_usa_pivot.set_index('job_posted_month', inplace = True)

df_usa_pivot
```

```python
df_usa_pivot.drop(columns = 'Month Nos', inplace = True)

df_usa_pivot
```

```python
df_usa_pivot.plot(kind = 'line')
```

```python
top_jobs = df_usa['job_title_short'].value_counts().head(3)

top_jobs = top_jobs.index.to_list()

top_jobs
```

```python
df_usa_pivot[top_jobs].plot(kind = 'line')
plt.ylabel('Number of Jobs')
plt.xlabel('2023')
plt.title('Monthly Job Posting For Top Data Jobs in the US')
plt.legend()
plt.show()
```

## Merging Data

```python
pd.read_csv('https://lukeb.co/software_csv')
```

```python
us_dev_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col = 'job_posted_month')

us_dev_pivot
```

```python
tech_jobs = df_usa_pivot.merge(us_dev_pivot, on = 'job_posted_month')

tech_jobs
```

```python
top_tech = (tech_jobs
    .sum()
    .sort_values(ascending = False)
    .head()
    .index.to_list()
)
top_tech
```

```python
tech_jobs[top_tech].plot(kind = 'line')
plt.ylabel('Number of Jobs')
plt.ylim(0, 20000)
plt.xlabel('2023')
plt.title('Monthly Job Posting For Top Tech Jobs in the US')
plt.legend()
plt.show()
```

## Pandas Concatenate

We can use the concat function to add multiple dataframes together. But unlike merge, it does so along a particular axis.
For this, we need to generate data into a dataframe from our existing dataset, so we can try the concat function.

```python
df['job_country'].value_counts().head(20) # Checking to see what country to work on.
```

```python
df_uk = df[df['job_country'] == 'United Kingdom'].copy()

df_uk.sample(4)
```

```python
df_uk['job_posted_month'] = df_uk['job_posted_date'].dt.strftime('%b')  # Creating a new column to store the month of the job posting.

df_uk.sample(4)
```

```python
# to create a dictionary so we can separate use the months as keys and their df as values. I'll illustrate below.

# dict_months = {
#     'Jan': df_jan,
#     'Feb': df_feb,
#     'Mar': df_mar,
#     'Apr': df_apr,
#     'May': df_may,
#     'Jun': df_jun,
#     'Jul': df_jul,
#     'Aug': df_aug,
#     'Sep': df_sep,
#     'Oct': df_oct,
#     'Nov': df_nov,
#     'Dec': df_dec,
#}

months = df_uk['job_posted_month'].unique() # This pulls out all the unique entries in said column

# Now I'll use dict comprehension.

{month: month for month in months}
```

```python
dict_month = {month:df_uk[df_uk['job_posted_month'] == month] for month in months}  # Now we've assigned the df to the keys and assigned the dict to a variable.
```

```python
dict_month['Jun'].sample(4)  # Now we can call each month in a df like this.
```

```python
# We now have dataframes to concatenate finally lol.
```

2023 Quater 1 UK Tech Job Postings Data

```python
uk_qt1 = pd.concat([
dict_month['Jan'], 
dict_month['Feb'], 
dict_month['Mar']
],
ignore_index = True)

uk_qt1
```

```python
uk_qt1['job_posted_month'].value_counts().plot(kind = 'bar')
```
